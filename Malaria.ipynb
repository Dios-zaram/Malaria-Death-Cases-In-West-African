```python
import numpy as np
import pandas as pd
WA_population=pd.DataFrame(pd.read_csv("C:/Program Files/PostgreSQL/15/data/data_copy/Western Africa.csv"))
Malaria=pd.DataFrame(pd.read_csv('C:/Program Files/PostgreSQL/15/data/data_copy/malaria dataset.csv'))
```

```python
WA_countries=pd.DataFrame(pd.read_csv('C:/Program Files/PostgreSQL/15/data/data_copy/world-data-2023.csv'))
WA_countries
```

```python
#Cleaning to only include West African countries
WA_countries=WA_countries[WA_countries['Country'].isin(['Benin','Burkina Faso','Ivory Coast','Cape Verde','Cameroon','The Gambia',
                                      'Ghana','Guinea','Guinea-Bissau','Liberia','Mali','Mauritania','Niger','Nigeria',
                                      'Senegal','Sierra Leone','Togo'])]

#Right Data Type
WA_countries_copy = WA_countries.copy()
WA_countries_copy['Country'] = WA_countries_copy['Country'].astype(str)
WA_countries_copy['Population'] = WA_countries_copy['Population'].astype(int)
WA_countries_copy['Urban_population'] = WA_countries_copy['Urban_population'].astype(int)
#print(WA_countries.dtypes)

#Updating Column Dataset


WA_countries_copy.loc[62,'Country']='Gambia'
WA_countries_copy.loc[28,'Country'] ='Cote dIvoire(Ivory Coast)'

#Droping duplicates
WA_countries_copy.drop_duplicates(subset='Country')
```

```python
WA_population=pd.DataFrame(WA_population)
WA_population_copy=WA_population.copy()

#Drop null values and duplicates
WA_population_copy=WA_population_copy.dropna()
WA_population_copy=WA_population_copy.drop_duplicates(subset=(['Country','Year','Population']))

#Right Datatype
WA_population_copy['Population']=WA_population_copy['Population'].astype(int)
WA_population_copy['Year']=WA_population_copy['Year'].astype(int)
WA_population_copy['Country']=WA_population_copy['Country'].astype(str)
WA_population_copy.dtypes

#update change
WA_population_copy.loc[ WA_population_copy['Country'] == 'Gambia, The', 'Country'] = 'Gambia'
WA_population_copy.loc[ WA_population_copy['Country'] == 'Cabo Verde','Country'] = 'Cape Verde'

WA_population_copy[WA_population_copy['Country'].isin(['Cape Verde', 'Gambia'])]

WA_population_copy
```

```python
#Make copy
Malaria_copy = Malaria.copy()

#Only West African Countries
Malaria_copy=Malaria[Malaria['location_name'].isin(['Benin','Burkina Faso','Ivory Coast','Cape Verde','Cameroon','The Gambia',
                                      'Ghana','Guinea','Guinea-Bissau','Liberia','Mali','Mauritania','Niger','Nigeria',
                                      'Senegal','Sierra Leone','Togo'])]

#Drop irrelevant columns
Malaria_copy=Malaria_copy.drop(['location_id', 'measure_id','sex_id','sex_name','metric_id','age_id','age_name'], axis=1)

#Correct Datatypes
change = ['measure_name', 'location_name', 'cause_name', 'metric_name', 'year', 'val', 'upper', 'lower']
data_types = {'measure_name': str, 'location_name': str, 'cause_name': str, 'metric_name': str, 'year': int, 'val': float, 'upper': float, 'lower': float}
Malaria_copy[change] = Malaria_copy[change].astype(data_types)

#Update
Malaria_copy=Malaria_copy[Malaria_copy['measure_name']=='Deaths']
Malaria_copy=Malaria_copy[Malaria_copy['metric_name']=='Number']
Malaria_copy=Malaria_copy[Malaria_copy['year']>=2015]

Malaria_copy
```

```python
#ANALYSIS
round(Malaria_copy.groupby(['location_name'])['val'].agg(np.sum))
we=Malaria_copy[Malaria_copy['year']==2019]
round(we.groupby(['location_name','year'])[['val']].agg(np.sum))

```
